```python
from sklearn.manifold import TSNE
import numpy as np
import time
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go

from cvjson.cvj import CVJ
from resvm import get_data
```

```python
def threeD_data(data, labels):
    df = pd.DataFrame(data)
    df.columns = ["x", "y", "z"]

    df['labels'] = labels

    return df

def twoD_data(data, labels):
    df = pd.DataFrame(data)
    df.columns = ["x", "y"]

    df['labels'] = labels

    return df

def tsne_3d_kmeans(kmeans_clf, data, layout=None, file_name=None):

    labels = kmeans_clf.fit(data)

    df['labels'] = labels.labels_

    traces = []
    for label in labels:
        traces.append(go.Scatter3d(x=df[df['labels'] == label]['x'], y=df[df['labels'] == label]['y'], z=df[df['labels'] == label]['z'],
                        mode='markers', name="Cluster_" + label))

    data = go.Data(traces)

    if layout != None:
        data = go.Figure(data=data, layout=layout)

    if file_name == None:
        if layout != None:
            file_name = layout.title
        else:
            file_name = "tsne_3d_kmeans"

    pyo.plot(data, filename=file_name +".html")

    return traces

def tsne_2d_kmeans(kmeans_clf, data, layout=None, file_name=None):

    labels = kmeans_clf.fit(data)

    df['labels'] = labels.labels_

    traces = []
    for label in labels:
        traces.append(go.Scatter3d(x=df[df['labels'] == label]['x'], y=df[df['labels'] == label]['y'],
                        mode='markers', name="Cluster_" +label))

    data = go.Data(traces)

    if layout != None:
        data = go.Figure(data=data, layout=layout)

    if file_name == None:
        if layout != None:
            file_name = layout.title
        else:
            file_name = "tsne_2d_kmeans"

    pyo.plot(data, filename=file_name +".html")

    return traces

def tsne_3d(data, cvj_obj, layout=None, file_name=None):

    labels = df['labels'].unique()

    traces = []
    for label in labels:
        traces.append(go.Scatter3d(x=df[df['labels'] == label]['x'], y=df[df['labels'] == label]['y'],
                        mode='markers', name=cvj_obj.get_class_id_2_name(label)))

    data = go.Data(traces)

    if layout != None:
        data = go.Figure(data=data, layout=layout)

    if file_name == None:
        if layout != None:
            file_name = layout.title
        else:
            file_name = "tsne_3d"


    pyo.plot(data, filename=file_name + ".html")

    return traces
```
